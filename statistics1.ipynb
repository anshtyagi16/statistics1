Mean, Median, Mode

```python
import statistics

data = [10, 20, 30, 40, 50]

print("Mean:", statistics.mean(data))
print("Median:", statistics.median(data))
print("Mode:", statistics.mode(data))
```

Variance & Standard Deviation

```python
import numpy as np

data = [10, 20, 30, 40, 50]

print("Variance:", np.var(data))
print("Standard Deviation:", np.std(data))
```

Random & Stratified Sampling

```python
import pandas as pd

data = pd.DataFrame({"Marks":[45,60,70,85,90], "Grade":["C","B","B","A","A"]})

random_sample = data.sample(3)
print(random_sample)
```

Range Function

```python
def find_range(data):
    return max(data) - min(data)

print(find_range([10, 20, 30, 40]))
```

Histogram (Skewness)

```python
import matplotlib.pyplot as plt

data = [10,12,13,15,20,30,45]

plt.hist(data)
plt.show()
```

Skewness & Kurtosis

```python
from scipy.stats import skew, kurtosis

data = [10,12,13,15,20,30,45]

print("Skewness:", skew(data))
print("Kurtosis:", kurtosis(data))
```

Covariance

```python
import numpy as np

x = [10,20,30]
y = [40,50,60]

print(np.cov(x, y))
```

Correlation

```python
print(np.corrcoef(x, y))
```

Scatter Plot

```python
plt.scatter(x, y)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

Descriptive Statistics using pandas

```python
import pandas as pd

df = pd.DataFrame({"Marks":[45,55,65,75,85]})
print(df.describe())
```

Boxplot

```python
plt.boxplot(df["Marks"])
plt.show()
```

IQR

```python
Q1 = df["Marks"].quantile(0.25)
Q3 = df["Marks"].quantile(0.75)
print("IQR:", Q3 - Q1)
```

Z-Score Normalization

```python
from scipy.stats import zscore

df["Zscore"] = zscore(df["Marks"])
print(df)
```

Pearson & Spearman Correlation

```python
from scipy.stats import pearsonr, spearmanr

print(pearsonr(x, y))
print(spearmanr(x, y))
```
